```python
import pandas as pd
import matplotlib.pyplot as plt


df1 = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_2583778.csv", skiprows=4)
df2 = pd.read_csv("Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_2583778.csv")
df3 = pd.read_csv("Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_2583778.csv")


print(df1.columns)


df1 = df1.drop(columns=['Indicator Name', 'Indicator Code', df1.columns[-1]]).dropna()


latest_year = df1.columns[-1]


top_10_countries = df1[['Country Name', latest_year]].sort_values(by=latest_year, ascending=False).head(10)


plt.figure(figsize=(12, 8))
plt.bar(top_10_countries['Country Name'], top_10_countries[latest_year], color='skyblue')
plt.title('Top 10 Most Populous Countries in ' + latest_year)
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
